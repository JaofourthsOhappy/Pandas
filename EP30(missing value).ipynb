df.isnull()
    True ข้อมูลไม่ครบ
    False ข้อมูลครบ
df.isnull().any()   หาคอลัม์ใดที่มีค่าว่าง
df.isnull().sum()   มีค่าว่างกี่เเถว

df.notnull()
    True ข้อมูลครบ
    False ข้อมูลไม่ครบ
df.notnull().any()   หาคอลัม์ใดที่ข้อมูลครบ
df.notnull().sum()   มีข้อมูลครบถ้วนกี่เเถว
