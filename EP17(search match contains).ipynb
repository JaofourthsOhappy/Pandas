match(string)  = ค้นหาคำโดยเริ่มจากอักขระตัวเเรก
contains(string) = ค้นหาคำไม่จำเป็นต้องเริ่มจากอักขระตัวเเรก

```python
import pandas as pd
df = pd.read_excel("dataupdate.xlsx","weather",index_col="Day")
```

```python
df[df.Event.str.match("ฝนตก")]
```

```python
df[df.Event.str.contains("ก")]
```
